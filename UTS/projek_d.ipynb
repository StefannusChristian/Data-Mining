```python
df.dropna(inplace=True)
df.shape
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

df['StockCode'] = df['StockCode'].astype(str)
encoded_stock_code = pd.DataFrame(label_encoder.fit_transform(df['StockCode']))
encoded_description = pd.DataFrame(label_encoder.fit_transform(df['Description']))
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

df['Quantity'] = scaler.fit_transform(df[['Quantity']])
df['UnitPrice'] = scaler.fit_transform(df[['UnitPrice']])
```

```python
X = pd.concat([df['Quantity'].reset_index(drop=True), df['UnitPrice'].reset_index(drop=True), encoded_stock_code.reset_index(drop=True), encoded_description.reset_index(drop=True)], axis=1)
X.isnull().sum()
```

```python
X_baru = X.groupby("Encoded_Description").mean().reset_index()
X_baru
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=5, random_state=0).fit(X_baru)
labels = kmeans.labels_
X_baru['Cluster'] = labels
# df = df.sort_values("Cluster")
# print(df.shape)
# for i in range(5):
#     print(i,(len(df[df["Cluster"] == 0])))
X_baru.drop(["Quantity","Encoded_StockCode","UnitPrice"],axis=1)
```
